```python
import pandas as pd

alimentos_data = pd.read_csv('C:/Users/gu-gu/OneDrive/Desktop/Big Data/3 Semestre/PI 3/elt nutricao.csv')
alimentos_nomes = alimentos_data['food']

alimentos_data.drop('food',axis=1,inplace=True)
df = alimentos_data.to_numpy()
alimentos_data
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import datasets
from sklearn_som.som import SOM

# Crio uma SOM com 3x1 neurônios (3 clusters)
som = SOM(m=3, n=1, dim=len(df[0]), random_state=1234)

# Busca de agrupamentos
som.fit(df)

# Associo cada país com o grupo respectivo
predictions = som.predict(df)
```

```python
dic = {}
for i,p in enumerate(predictions):
  if p in dic.keys():
    dic[p].append(alimentos_nomes.iloc[i])
  else: dic[p] = [alimentos_nomes.iloc[i]]
```

```python
for i in dic.keys():
  lista = dic[i]
  print(sorted(lista))
  print('---------------------------------')
```

```python
df2 = alimentos_data
df2['food'] = alimentos_nomes
df2['cluster'] = predictions
```
